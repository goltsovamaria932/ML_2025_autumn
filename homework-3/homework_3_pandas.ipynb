```python
import pandas as pd
```

```python
pass_data = pd.read_csv('chipotle.tsv', sep='\t')
pass_data
```

```python
string_columns = pass_data.select_dtypes(include=['object']).columns
numeric_columns = pass_data.select_dtypes(include=['number']).columns

pass_data[string_columns] = pass_data[string_columns].fillna('')

pass_data[numeric_columns] = pass_data[numeric_columns].fillna(0)

pass_data.isnull().sum()
```

```python
pass_data
```

```python
def convert_price(price_str):
  return float(price_str.replace('$', ''))

pass_data['price_numeric'] = pass_data['item_price'].apply(convert_price)

pass_data
```

Столбец с ценами имеет строковый тип, так как там используется значок доллара.

```python
print('Среднее:', pass_data['price_numeric'].mean())
print('Минимальное:', pass_data['price_numeric'].min())
print('Максимальное:', pass_data['price_numeric'].max())

pass_data = pass_data.drop('item_price', axis=1)
pass_data
```

```python
unique_items = pass_data.drop_duplicates('item_name')

cheapest_items = unique_items.nsmallest(5, 'price_numeric')[['item_name', 'price_numeric']]

expensive_items = unique_items.nlargest(5, 'price_numeric')[['item_name', 'price_numeric']]

print("5 самых дешёвых товаров:")
for idx, row in cheapest_items.iterrows():
    print(f"{row['item_name']}: ${row['price_numeric']:.2f}")

print("\n5 самых дорогих товаров:")
for idx, row in expensive_items.iterrows():
    print(f"{row['item_name']}: ${row['price_numeric']:.2f}")
```

```python
chicken_bowl_orders = pass_data[pass_data['item_name'] == 'Chicken Bowl'].groupby('order_id')['quantity'].sum()

count_orders = len(chicken_bowl_orders[chicken_bowl_orders > 1])

print(f"Количество заказов с больше чем 1 Chicken Bowl: {count_orders}")
```

```python
order_summary = pass_data.groupby('order_id').agg(
    total_amount=('price_numeric', 'sum'),
    total_quantity=('quantity', 'sum'),
).reset_index()

avg_receipt = order_summary['total_amount'].mean()
avg_quantity = order_summary['total_quantity'].mean()

print(f"Средний чек: ${avg_receipt:.2f}")
print(f"Среднее количество товаров: {avg_quantity:.0f} шт.")

print(order_summary['total_amount'].describe())
```

```python
order_quantities = pass_data.groupby('order_id')['quantity'].sum()

single_item_orders = order_quantities[order_quantities == 1]

count_single_item_orders = len(single_item_orders)

print(f"Количество заказов с ровно 1 товаром: {count_single_item_orders}")
```

```python
popularity = pass_data['item_name'].value_counts()

most_popular_item = popularity.head(1)

print("Самый популярный товар:")
print(f"{most_popular_item.index[0]}: {most_popular_item.values[0]} продаж")
```

```python
burrito_items = pass_data[pass_data['item_name'].str.contains('Burrito', na=False)]

burrito_types = burrito_items['item_name'].unique()

print("Виды BURRITO в меню:")
for i, burrito in enumerate(sorted(burrito_types), 1):
    print(f"{i}. {burrito}")

burrito_popularity = burrito_items['item_name'].value_counts()

print(f"\nСамый популярный BURRITO:")
most_popular = burrito_popularity.head(1)
print(f"{most_popular.index[0]}: {most_popular.values[0]} продаж")

burrito_prices = burrito_items.groupby('item_name')['price_numeric'].agg(['max', 'min', 'mean']).round(2)

print(f"\nСамый дорогой BURRITO:")
most_expensive = burrito_prices.nlargest(1, 'max')
burrito_name = most_expensive.index[0]
price = most_expensive['max'].values[0]
print(f"{burrito_name}: ${price}")
```

```python
order_totals = pass_data.groupby('order_id')['price_numeric'].sum().reset_index()
order_totals = order_totals.rename(columns={'price_numeric': 'order_total'})

pass_data_with_totals = pass_data.merge(order_totals, on='order_id')

pass_data_with_totals['price_ratio'] = (pass_data_with_totals['price_numeric'] / pass_data_with_totals['order_total']) * 100

high_ratio_items = pass_data_with_totals[pass_data_with_totals['price_ratio'] > 40]

orders_with_high_ratio = high_ratio_items['order_id'].nunique()

print(f"Количество заказов, где есть товар стоимостью более 40% от суммы чека: {orders_with_high_ratio}")
```

```python
original_total = pass_data['price_numeric'].sum()

diet_coke_mask = pass_data['choice_description'].str.contains('Diet Coke', na=False)
diet_coke_original = pass_data[diet_coke_mask].copy()

price_difference_per_unit = 1.35 - 1.25

pass_data.loc[diet_coke_mask, 'price_numeric'] += diet_coke_original['quantity'] * price_difference_per_unit

new_total = pass_data['price_numeric'].sum()

diet_coke_new = pass_data[diet_coke_mask]
diet_coke_original_total = diet_coke_original['price_numeric'].sum()
diet_coke_new_total = diet_coke_new['price_numeric'].sum()

diet_coke_revenue_increase = diet_coke_new_total - diet_coke_original_total
diet_coke_percentage_increase = (diet_coke_revenue_increase / diet_coke_original_total) * 100

total_revenue_increase = new_total - original_total
total_percentage_increase = (total_revenue_increase / original_total) * 100

print(f"Исходная общая выручка: ${original_total:.2f}")
print(f"Новая общая выручка: ${new_total:.2f}")
print(f"Общее увеличение выручки: ${total_revenue_increase:.2f}")
print(f"Общий процент увеличения: {total_percentage_increase:.4f}%")

print(f"Исходная выручка от Diet Coke: ${diet_coke_original_total:.2f}")
print(f"Новая выручка от Diet Coke: ${diet_coke_new_total:.2f}")
print(f"Увеличение выручки от Diet Coke: ${diet_coke_revenue_increase:.2f}")
print(f"Процент увеличения выручки от Diet Coke: {diet_coke_percentage_increase:.2f}%")
```

```python
matrix = [
    ['Убийственный класс', '(комедия/боевик)'],
    ['Стальной Алхимик: Братство', '(боевик)'],
    ['Саики Кусуо', '(комедия)'],
    ['Твоя апрельская ложь', '(романтика/драма)'],
    ['Нитидзё', '(комедия)'],
    ['Steins;Gate', '(фантастика/триллер)'],
    ['Хоримия', '(романтика)'],
    ['Моя геройская академия', '(комедия/боевик)'],
    ['Fate/Zero', '(триллер/боевик)'],
    ['Корзинка фруктов', '(романтика)'],
    ['Плутон', '(триллер/боевик)'],
    ['Галактика Татами', '(комедия/фантастика)'],
    ['Атака титанов', '(боевик/триллер)'],
    ['Ковбой Бибоп', '(фантастика/боевик)'],
    ['Шова Генроку Ракуго Синдзю', '(драма)']
]

data_dict = {
    'index': [row[0] for row in matrix],
    'column1': [row[1] for row in matrix],
    'column2': [None] * len(matrix)
}

pass_data_film = pd.DataFrame(data_dict)

pass_data_film = pass_data_film.set_index('index')

pass_data_film
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Левый график - суммы заказов
pass_data.groupby('order_id')['price_numeric'].sum().hist(bins=30, ax=ax1, color='lightblue')
ax1.set_title('Распределение сумм заказов')
ax1.set_xlabel('Сумма ($)')
ax1.set_ylabel('Количество заказов')

# Правый график - средние цены товаров
pass_data.groupby('item_name')['price_numeric'].mean().hist(bins=20, ax=ax2, color='lightgreen')
ax2.set_title('Распределение средних цен товаров')
ax2.set_xlabel('Цена ($)')
ax2.set_ylabel('Количество товаров')

plt.tight_layout()
plt.show()
```

```python
pass_data_sorted = pass_data.sort_values('order_id')
daily_totals = pass_data_sorted.groupby('order_id')['price_numeric'].sum()

plt.figure(figsize=(12, 6))
plt.plot(daily_totals.index, daily_totals.values, alpha=0.7, linewidth=1)
plt.title('Суммы покупок по заказам (в хронологическом порядке)')
plt.xlabel('Номер заказа')
plt.ylabel('Сумма покупки ($)')
plt.grid(True, alpha=0.3)
plt.show()
```
